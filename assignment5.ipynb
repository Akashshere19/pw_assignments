11.What is memory management in Python?

Python's memory management components provides efficient and effective utilization of memory resources throughout the execution of Python programs.
The Python memory manager internally ensures the efficient allocation and deallocation of this memory.

Python manages memory allocation in two primary ways − Stack and Heap.

In static memory allocation, memory is allocated at compile time and stored in the stack.It operates on a Last-In-First-Out (LIFO) basis.

Dynamic memory allocation occurs at runtime for objects and data structures of non-primitive types.

Garbage Collection in Python is the process of automatically freeing up memory that is no longer in use by objects

```python
# 18.How would you implement a basic logging setup that logs to a file with rotation after 1MB
from logging.handlers import RotatingFileHandler
logger = logging.getLogger('my_logger')
logger.setLevel(logging.DEBUG)

handler = RotatingFileHandler('error.log', maxBytes=1024*1024, backupCount=5)
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)
logger.addHandler(handler)
for i in range(10000):
    logger.debug(f"This is a very large log message {i} " + "X" * 500)

```
