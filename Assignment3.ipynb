6. Explain the concept of generators in Python and how they are defined.

generator is special function or expression that concept related to generate iterable items but one element at time.It use yield keyword for generate output.It does not occupied space total no of iterable items.so it generate one element from it at a time.






```python
def fruits(lst):
  for itr in lst:
    yield itr

obj = fruits(['apple','mango','graphs','cherry'])
print(next(obj))
print(next(obj))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def checkPrime(num):
  prime = True
  if num <= 1:
    prime = False
  for j in range(2,int(num**0.5)+1,2):
        if num % j ==0:
          prime = False
          break
  return prime
result = []
for num in range(1,200):
  if checkPrime(num):
        result.append(num)
print(result)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file(path):
  with open(path,'r') as file:
    # r = file.read()
    for r in file:
      yield r
    file.close()
flt = read_file('/content/social_setup')
print(next(flt))
print(next(flt))
print(next(flt))
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
nameAge = [('ankit',32),('janvi',29),('kartik',22),('bhavani',27),('mohan',35)]
sortAge = sorted(nameAge,key=lambda age:age[1])
print(list(sortAge))
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

myString = 'apple goes umbrella came!'
vowel = 'aeiou'
filterString = filter(lambda s: s not in vowel ,myString)
print(''.join(list(filterString)))
```
