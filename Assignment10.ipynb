Question 1: What is hypothesis testing in statistics?

It is is a statistical method to test a claim about a population using sample data.
It involves defining a null hypothesis (a statement of no effect) and an alternative hypothesis (a statement claiming a specific effect), collecting sample data, and then determining if the sample evidence is strong enough to reject the null hypothesis in favor of the alternative.


Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

states that there is no significant effect, relationship, or difference between variables, while the alternative hypothesis (H₁ or Hₐ) proposes that a statistically significant effect, relationship, or difference does exist.

The main difference is that the null hypothesis is a statement of the default or baseline assumption of no effect, which the researcher tries to disprove, whereas the alternative hypothesis is the researcher's prediction or claim that they are trying to prove



Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.


The significance level (alpha, α) is the pre-set probability of incorrectly rejecting a true null hypothesis, representing the maximum risk of a Type I error a researcher is willing to take.

Question 4: What are Type I and Type II errors? Give examples of each.

* Type I and Type II errors are mistakes in hypothesis testing.
* A Type I error is a false positive (incorrectly rejecting a true null hypothesis), like concluding a drug is effective when it's not.
* A Type II error is a false negative (failing to reject a false null hypothesis), such as failing to identify a drug as effective when it actually is

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

* The main difference is that a Z-test is used when the population standard deviation is known and the sample size is large (typically n ≥ 30), while a T-test is used when the population standard deviation is unknown and the sample size is small (typically n < 30).

* Z-test assumes a normal distribution with known variance, whereas the T-test uses the t-distribution to account for the uncertainty of an unknown standard deviation in smaller samples

Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
n = 10
p = 0.5

x = np.arange(0,n+1)
probabilities = binom.pmf(x, n, p)
plt.figure(figsize=(10, 6))
plt.bar(x, probabilities, color='skyblue', edgecolor='black')
plt.xlabel("Number of Successes (k)")
plt.ylabel("Probability P(X=k)")
plt.title(f"Binomial Distribution Histogram (n={n}, p={p})")
plt.xticks(x)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.

import math
import numpy as np
from statsmodels.stats.weightstats import ztest
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


mu0 = 50
n = len(sample_data)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

z_stat = (sample_mean - mu0) / (sample_std / n**0.5)

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean = {sample_mean:.4f}")
print(f"Sample Std Dev = {sample_std:.4f}")
print(f"Z-statistic = {z_stat:.4f}")
print(f"P-value = {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject Null Hypothesis (H0): Sample mean is significantly different from 50.")
else:
    print("Fail to Reject Null Hypothesis (H0): No significant difference from 50.")

```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt

mu = 100
sigma = 15
n = 1000
np.random.seed(42)
data = np.random.normal(loc=mu, scale=sigma, size=n)


sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

sem = sample_std / np.sqrt(n)


z_score = 1.96

margin_of_error = z_score * sem

confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({confidence_interval_lower:.2f}, {confidence_interval_upper:.2f})")

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', label='Simulated Data')

plt.axvline(sample_mean, color='red', linestyle='solid', linewidth=2, label='Sample Mean')
plt.axvline(confidence_interval_lower, color='green', linestyle='dashed', linewidth=2, label='95% CI Lower Bound')
plt.axvline(confidence_interval_upper, color='green', linestyle='dashed', linewidth=2, label='95% CI Upper Bound')

plt.title('Simulated Normal Distribution Data with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_visualize_zscores(data):

    data = np.array(data)

    mean = np.mean(data)
    std_dev = np.std(data)

    z_scores = (data - mean) / std_dev

    print(f"Original Data: {data}\n")
    print(f"Mean of Data: {mean:.4f}")
    print(f"Standard Deviation of Data: {std_dev:.4f}\n")
    print(f"Calculated Z-Scores: {np.round(z_scores, 4)}\n")

    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=20, edgecolor='black', alpha=0.7)
    plt.title('Histogram of Standardized Data (Z-Scores)', fontsize=16)
    plt.xlabel('Z-Score', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z-score = 0)')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

z_scores_output = calculate_and_visualize_zscores(sample_data)
```
